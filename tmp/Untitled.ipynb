```python
from matplotlib import pyplot as plt
```

```python
x1 = 1
x2 = -1

v1 = 0
v2 = 0

a1 = 0
a2 = 0

dt = 0.00001

m = 2
k = 1
```

```python
a_x1 = []
a_x2 = []

for i in range(3000000):
    x1 += dt * v1
    x2 += dt * v2
    
    v1 += a1 * dt
    v2 += a2 * dt
    
    a1 = (-3 * x1 + 2 * x2) * k / m
    a2 = (2 * x1 - 5 * x2) * k / m

    a_x1.append(x1)
    a_x2.append(x2)

time = [i * dt for i in range(3000000)]
```

```python
plt.plot(time, a_x1)
plt.xticks([i for i in range(0, 31)])
```

```python
plt.plot(time, a_x2)
```
